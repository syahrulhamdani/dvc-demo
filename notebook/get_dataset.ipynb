# Get Dataset

We will use this notebook the get the dataset for the workshop. We will use the datasets provided by `deepchecks`.

> [deepchecks](https://deepchecks.com/) is a ML validation tools to build test suites for ML model and data.

## Phishing Dataset

This dataset is based on the [project](https://github.com/Rohith-2/url_classification_dl) by [Rohith Ramakrishnan](https://www.linkedin.com/in/rohith-ramakrishnan-54094a1a0/) licensed under [CCO 1.0](https://creativecommons.org/publicdomain/zero/1.0/).

Some information regarding the dataset:
* 11.35K samples
* 25 columns
* consists of real numbers and categorical features
* target variable is boolean

```
phishing_data = phishing.load_data(data_format="Dataframe", as_train_test=False)
phishing_data.info()
```

```
phishing_data.head()
```

```
phishing_data.to_csv("dvc-demo/data/raw/phishing.csv", index=False)
```

## Airbnb Dataset

This dataset was orginally adapted from the open Kaggle dataset https://www.kaggle.com/datasets/dgomonov/new-york-city-airbnb-open-data downloaded from the Airbnb data repository http://insideairbnb.com/get-the-data. The dataset is licensed under the [Creative Commons Attribution 4.0 International License](https://creativecommons.org/licenses/by/4.0/).

Some information regarding the dataset:
* 47.3K samples
* 9 columns
* consists of numerical and categorical features
* target values ranging from 31 to 795

```
airbnb_data = airbnb.load_data(data_format="Dataframe", as_train_test=False)
airbnb_data.info()
```

```
airbnb_data.head()
```

```
airbnb_data.to_csv("dvc-demo/data/raw/airbnb.csv", index=False)
```
